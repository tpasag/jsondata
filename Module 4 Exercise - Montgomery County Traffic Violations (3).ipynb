```python
#create variable for where you want to extract your data
violations = vio_data['data']
```

```python
type(violations)
```

```python
#create dictionary so that each item will be a column header when you load the data again
#rerun this code again if you change your for loop below 
vio_dict={
'date': [],
'time':[],
'agency':[],
'subagency':[],
'description':[],
'location':[],
'latitude':[],
'longitude':[],
'vehicletype':[],
'vehicleyear':[],
'vehiclemake':[],
'vehiclemodel':[],
'vehiclecolor':[],
'violationtype':[],
'race':[],
'gender':[],
'driverstate':[],
'driverslicensestate':[],
'arrestype':[]
}
```

```python
#for loop means for item in list 
#get all these items in these index positions
#if you make changes to the for loop, you must reempty out the list by 
#...rerunning the code box above!
for violation in violations:
    vio_dict['date'].append(violation[9])
    vio_dict['time'].append(violation[10])
    vio_dict['agency'].append(violation[11])
    vio_dict['subagency'].append(violation[12])
    vio_dict['description'].append(violation[13])
    vio_dict['location'].append(violation[14])
    vio_dict['latitude'].append(violation[15])
    vio_dict['longitude'].append(violation[16])
    vio_dict['vehicletype'].append(violation[28])
    vio_dict['vehicleyear'].append(violation[29])
    vio_dict['vehiclemake'].append(violation[30])
    vio_dict['vehiclemodel'].append(violation[31])
    vio_dict['vehiclecolor'].append(violation[32])
    vio_dict['violationtype'].append(violation[33])
    vio_dict['race'].append(violation[37])
    vio_dict['gender'].append(violation[38])
    vio_dict['driverstate'].append(violation[40])
    vio_dict['driverslicensestate'].append(violation[41])
    vio_dict['arrestype'].append(violation[42])
```
